```python
from braindecode.datasets.moabb import MOABBDataset
dataset = MOABBDataset(dataset_name="BNCI2014001", subject_ids=[1,2,3,4,5,6,7,8,9])
```

pre-processing the data

```python
from braindecode.datautil.preprocess import (
    exponential_moving_standardize, preprocess, Preprocessor)

low_cut_hz = 4.  # low cut frequency for filtering
high_cut_hz = 38.  # high cut frequency for filtering
# Parameters for exponential moving standardization
factor_new = 1e-3
init_block_size = 500

preprocessors = [
    Preprocessor('pick_types', eeg=True, meg=False, stim=False),  # Keep EEG sensors
    Preprocessor(lambda x: x * 1e6),  # Convert from V to uV
    Preprocessor('filter', l_freq=low_cut_hz, h_freq=high_cut_hz),  # Bandpass filter
    Preprocessor(exponential_moving_standardize,  # Exponential moving standardization
                 factor_new=factor_new, init_block_size=init_block_size)
]

# Transform the data
preprocess(dataset, preprocessors)
```

cut windows


```python
import numpy as np
from braindecode.datautil.windowers import create_windows_from_events

trial_start_offset_seconds = -0.5
# Extract sampling frequency, check that they are same in all datasets
sfreq = dataset.datasets[0].raw.info['sfreq']
assert all([ds.raw.info['sfreq'] == sfreq for ds in dataset.datasets])
# Calculate the trial start offset in samples.
trial_start_offset_samples = int(trial_start_offset_seconds * sfreq)

# Create windows using braindecode function for this. It needs parameters to define how
# trials should be used.
windows_dataset = create_windows_from_events(
    dataset,
    trial_start_offset_samples=trial_start_offset_samples,
    trial_stop_offset_samples=0,
    preload=True,
    window_size_samples = 500,
    window_stride_samples = 250,
)
```

method 2-splite trainging set and validation set

```python
# #先按照subject名划分出九个受试者的数据
# divide dataset into 9 datasets by subject_ids
splitted = windows_dataset.split('subject')
subject_1 = splitted['1'] #每个受试者的数据被切成2304条
subject_2 = splitted['2']
subject_3 = splitted['3']
subject_4 = splitted['4']
subject_5 = splitted['5']
subject_6 = splitted['6']
subject_7 = splitted['7']
subject_8 = splitted['8']
subject_9 = splitted['9']
per_sub_data_num = len(subject_1)
```

```python
#取出每个受试者数据的90%作为训练集
#use same ratio in every dataset to splite validation set 
import numpy as np
import random
train_set, valid_set = [], []
splite_ratio = 0.1
valid_num = int(per_sub_data_num*splite_ratio)
np.random.seed(1)
valid_idx = random.sample(list(range(per_sub_data_num)), valid_num) 
for i in range(per_sub_data_num):
  if i in valid_idx:
    valid_set.append(subject_1[i])
    valid_set.append(subject_2[i])
    valid_set.append(subject_3[i])
    valid_set.append(subject_4[i])
    valid_set.append(subject_5[i])
    valid_set.append(subject_6[i])
    valid_set.append(subject_7[i])
    valid_set.append(subject_8[i])
    valid_set.append(subject_9[i])
  else:
    train_set.append(subject_1[i])
    train_set.append(subject_2[i])
    train_set.append(subject_3[i])
    train_set.append(subject_4[i])
    train_set.append(subject_5[i])
    train_set.append(subject_6[i])
    train_set.append(subject_7[i])
    train_set.append(subject_8[i])
    train_set.append(subject_9[i])
  
```

```python
import numpy as np
train_X, train_y, valid_X, valid_y = [], [], [], []

for t in train_set:
  train_X.append(np.asarray(t[0])[[7, 11, 9], :])
  train_y.append(t[2][0])
for v in valid_set:
  valid_X.append(np.asarray(v[0])[[7, 11, 9], :])
  valid_y.append(v[2][0])

print(train_X[0].shape)

train_X, valid_X = np.asarray(train_X).reshape([-1, 3, 25, 20]).transpose([0, 2, 3, 1]), np.asarray(valid_X).reshape([-1, 3, 25, 20]).transpose([0, 2, 3, 1])
train_y, valid_y = np.asarray(train_y), np.asarray(valid_y)
print(train_X.shape)
print(valid_X.shape)
# for i in range(20):
  # print(valid_y[i])
```

```python
## import tf and check gpu is_available
import tensorflow as tf
assert tf.test.is_gpu_available()
```

```python
## build model
def build_model(inputs):

  # CONV_1
  C_1 = tf.keras.layers.Conv2D(32, (3, 3), padding="SAME", name="C_1", trainable=False)(inputs)
  B_1 = tf.keras.layers.BatchNormalization(name="B_1")(C_1)
  L_1 = tf.keras.layers.LeakyReLU()(B_1)
  # D_1 = tf.keras.layers.Dropout(rate=0.1)(L_1)
  # print(D_1.shape)

  # CONV_2
  C_2 = tf.keras.layers.Conv2D(32, (3, 3), padding="SAME", name="C_2", trainable=False)(L_1)
  B_2 = tf.keras.layers.BatchNormalization(name="B_2")(C_2)
  L_2 = tf.keras.layers.LeakyReLU()(B_2)
  # D_2 = tf.keras.layers.Dropout(rate=0.1)(L_2)
  # print(D_2.shape) #25,20,32
  
  # MAX_POOLING_1
  M_1 = tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=(2, 2), padding="SAME")(L_2) 
  print(M_1.shape) #13, 10, 32

  # CONV_3
  C_3 = tf.keras.layers.Conv2D(64, (3, 3), padding="SAME", name="C_3", trainable=False)(M_1)
  B_3 = tf.keras.layers.BatchNormalization(name="B_3")(C_3)
  L_3 = tf.keras.layers.LeakyReLU()(B_3)
  # D_3 = tf.keras.layers.Dropout(rate=0.1)(L_3)



  # CONV_4
  C_4 = tf.keras.layers.Conv2D(64, (3, 3), padding="SAME", name="C_4", trainable=False)(L_3)
  B_4 = tf.keras.layers.BatchNormalization(name="B_4")(C_4)
  L_4 = tf.keras.layers.LeakyReLU()(B_4)
  # D_4 = tf.keras.layers.Dropout(rate=0.1)(L_4)

  # MAX_POOLING_2
  M_2 = tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=(2, 2), padding="SAME")(L_4) 
  print(M_2.shape)#7,5,64

  #FC
  F_IN = tf.reshape(M_2, [-1, 7 * 5 * 64])
  F_1 = tf.keras.layers.Dense(128, name="F_1")(F_IN)
  F_1_B = tf.keras.layers.BatchNormalization(name="F_1_B")(F_1)
  F_1_L = tf.keras.layers.LeakyReLU()(F_1_B)
  F_1_D = tf.keras.layers.Dropout(rate=0.5)(F_1_L)

  # F_2 = tf.keras.layers.Dense(16, name="F_2")(F_1_L)
  # F_2_B = tf.keras.layers.BatchNormalization(name="F_2_B")(F_2)
  # F_2_L = tf.keras.layers.LeakyReLU()(F_2_B) 
 

  F_2 = tf.keras.layers.Dense(4, name="F_2")(F_1_D)
 
  logits = tf.nn.softmax(F_2)
  
  return logits
inputs = tf.keras.Input(shape=[25, 20, 3])
outputs = build_model(inputs)
model = tf.keras.Model(inputs, outputs)
model.compile(
  optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
  loss=tf.keras.losses.SparseCategoricalCrossentropy(),  #将数字编码转化成one-hot编码格式，然后对one-hot编码格式的数据（真实标签值）与预测出的标签值使用交叉熵损失函数
  metrics=[tf.keras.metrics.SparseCategoricalAccuracy()] #检查y_true中的值（本身就是index）与y_pred中的最大值对应的index是否相等，针对稀疏情况的多分类
)
model.load_weights(filepath="/content/drive/MyDrive/ckpt/high_gamma_pretrain_baseline/conv6net.h5", by_name=True)
```

```python
ckpt = tf.keras.callbacks.ModelCheckpoint(
    filepath="/content/drive/MyDrive/ckpt/bci2a_baseline/conv6net.h5",
    mointor="val_acc",
    save_best_only=True,
    save_weights_only=True,
    mode="max"
)
model.fit(
    x=train_X,
    y=train_y, 
    batch_size=64, #4.8.16.32
    epochs=120,  #20,30,,
    verbose="auto",
    validation_data=(valid_X, valid_y),
    callbacks=[ckpt],
    shuffle=True 
)
```
